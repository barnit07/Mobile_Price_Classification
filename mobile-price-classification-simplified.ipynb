```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/kaggle/input/mobile-price-classification/test.csv'
data = pd.read_csv(file_path)

# Define price range conditions based on feature thresholds
conditions = [
    (data['ram'] < 2000) & (data['battery_power'] < 900) & (data['int_memory'] < 16),  # Low-end phones
    (data['ram'] >= 2000) & (data['ram'] < 4000) & (data['battery_power'] >= 900) & (data['battery_power'] < 1400),  # Mid-range phones
    (data['ram'] >= 3000) & (data['battery_power'] >= 1400) & (data['int_memory'] >= 32), # High-end phones
    
]

# Assign price ranges based on the conditions
price_range = [0, 1, 2]  # 0: Low, 1: Medium, 2: High
data['price_range'] = np.select(conditions, price_range, default=1)  # Default to medium if no conditions match



# Print the results
print(data)
```

```python
X= data.iloc[:,1:-1].values # selecting all columns except last one i.e. price_range
y= data.iloc[:,-1].values # selecting last column i.e. price_range
```

```python
# splitting the dataset into test set and training set. Here, test set = 25%
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25, random_state=0);
```

```python
# Although feature scaling  is not important for the tree like model 
# for example Decision tree model and random forest model.

from sklearn.preprocessing import StandardScaler
sc= StandardScaler();
sc.fit_transform(X_train);
sc.transform(X_test);
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier= DecisionTreeClassifier(criterion='entropy',random_state=0) #creating the decision tree classification model.
classifier.fit(X_train, y_train); # training the model based on training sets
```

```python
print(classifier.predict([[2000,1,2,1,17,1,74,1,200,9,22,1300,1600,7000,18,12,12,1,1,1]])) #pridicting the mobile price classification
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)) # displaying the predicted result along with the the actual test set result to compare
```

```python
#evaluating the classification model

from sklearn.metrics import confusion_matrix, accuracy_score;
cm= confusion_matrix(y_test,y_pred)
print(cm);
print('Accuracy =',100*accuracy_score(y_test,y_pred),'%')
```
